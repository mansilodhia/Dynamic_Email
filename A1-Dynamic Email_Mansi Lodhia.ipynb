<br>
<h3>Building Email Template:</h3>

```python
# defining a personalized email function
def dynamic_email(name, category, hook):
    
    # Replacing unknown names with 'Valued Customer'
    if name == 'UNKNOWN':
        name = "Valued Customer"
    else:
        name = name.capitalize()
        
    # Capitalizing the category for a cleaner appearance in the email
    category = category.title()
        
    # Conditional logic to match HOOK TAG to respective Hook message
    if hook == "financing":
        hook_line = "FINANCING AT 0% INTEREST"
    elif hook == "return policy":
        hook_line = "A 30 DAY RETURN POLICY WITH NO STRINGS ATTACHED"
    elif hook == "in-home assembly":
        hook_line = "IN-HOME ASSEMBLY FOR JUST $10"
    elif hook == "free shipping":
        hook_line = "FREE SHIPPING"
    elif hook == "customization":
        hook_line = "DISCOUNTED PRICING ON TAILORING AND CUSTOMIZATIONS"
    # Closing with else clause and addition of hook line
    else:
        hook_line = "AN EXCLUSIVE DEAL JUST FOR YOU"
        
    # Building email structure with random function
    subject_line = (f"{category} {rand.choice(subjects)}")
    opening_line = (f"{rand.choice(opening)}{category}.")
    art_choice = rand.choice(art)
    promotion_line = (f"{rand.choice(promotion)}{hook_line}{rand.choice(promotion_code)}")
    
    # To print Subject line with formatting  
    print('-' * 50)
    print(f"SUBJECT: {subject_line.title()}")
    print('-' * 50)
    
    # To print Email Body with formatting
    print()
    print(f"Greetings {name}!\n")
    print(opening_line, "\n")
    print(promotion_line, "\n")
    print(art_choice, "\n")
    
    # To print Email Signature with formatting
    print(f"Best Wishes,\n{staff_name}\n{staff_title}\n\n")
    print('*' * 40, "\n\n")
```

<br>
<h3>Code to loop over the data and to develop personalized emails:</h3><br>

```python
# Looping through each customer in the customer_data list
for customer in customer_data:
    
    # Extracting customer_data with indent
    name = customer[0]
    category = customer[1]
    hook = customer[2]
    
    # Calling the dynamic_email function with the extracted data
    dynamic_email(name, category, hook)
```
